### Coding Exercise: Longest Palindromic Substring
### Given a string s, return the longest  palindromic substring in s.

### Example :

- Input: s = "pabad"
- Output: "aba"
# longest_pallindromic_substring problem Tabulation(bottom-up) approach [gap-strategy]:
### while filling T/F in 2-d dp_table through length-iteration or diagonal-iteration [Gap-strategy] instead of counting numbers of True in the dp table , here we have to find the specific maximum lengthed-diagonal with with atleast one True-value possbile ; and length corresponds to that diagonal will be the answer to longest_pallindromic_substring problem.

```python
# longest_pallindromic_substring problem Tabulation(bottom-up) approach [gap-strategy] [code implementation]:
# Palindromic Substrings problem Tabulation(bottom-up) approach [code implementation]:

def longestPalindrome(s): # this function specifically used to count total number of pallindromic substring possible  for given original string

    n = len(s) # length of given string 

    dp = [[0]*n for _ in range(n)] #2-d dp_table we construct of dimesnion (n*n)  intialised all cells with 0
    longest =''  # this is where we store the longest pallindromic substring possible from original-string
    
    for l in range(1,n+1): # length from 1 upto n
        for i in range(n-l+1): # i from 0 to n-l
            j = i+l - 1  # for each instance of i; j = i+l -1 
            if i==j:  # Longest-diagonal(median-diagonal): this diagonal where length equal to 1 ; means each and every char itself considered as substring 
                #and also considered Pallindrome certainly. 
                dp[i][j] = True
#last chars of curr-substring is equal then further check for substring in-betwwen these equal chars whether they're pallindromic substring or not (dp(i+1,j-1))
#// As we already computed value for substring in-between equal chars also  [that's the values already filled-in at smaller length diagonals ; length = curr-length-2] 
        # if its the case of adjacent equal chars (j==i+1) then no need to check pallindromic substring for further chars in-between. 
            elif s[i]==s[j] and (j==i+1 or dp[i+1][j-1]):  
                dp[i][j] = True
            else:
                dp[i][j] = False
            if dp[i][j]:  # if value in the dp_table at these current i & j indexes is True then : 
                longest = s[i:j+1]    # we will store the substring of original-string at indexes of dp_table (from current i to current j+1)
    return longest # and this is how we retriving or returning the last saved substring of original-string //its lastly saved to the 'longest' var when last True-value was at indexes dp[i][j].      

#example usage:
string = 'papad'   # longest pallindrome string possible 3 ; 'pap' or  'apa'
print(longestPalindrome(string)) # should be  3  ; 'apa' or 'pap'
st= 'abbabb'
print(longestPalindrome(st)) # should be  5 ; 'bbabb'     
```
