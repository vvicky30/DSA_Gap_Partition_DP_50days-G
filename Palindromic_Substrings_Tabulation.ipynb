# What is Gap-strategy : 
### Gap strategy actually also called as length-wise iteration ,
### in this strategy instead of traversing(iteratively) row-wise in the dp_table and filling-up values/solution of overlapping subproblems here we traverse(iteratively) diagonally in the dp_table.
### the longest diagonal possible on dp_table has such states i.e. (0,0) ,(1,1), (2,2) etc if i denotes row and j denotes colmn of dp_tble 
### the length of each cells on longest-diagonal possible on dp_table is 1 only   and length of each cell on further relatively small-diagonals will be 2,3,4 and 5 [as you can see on diagram below]
### similarly gap (or we can say difference b/w i & j) of each cells on longest-diagonal possible on dp_table is 0 (as 0-0= 0 ; 1-1=0 ; 2-2 = 0 etc ) and gap of each cell on further relatively small-diagonals will be 1,2,3 and 4.

# Palindromic Substrings problem Tabulation(bottom-up) approach:
### just like we did in recursion 7 memoisation approach where we filling-up 2-d dp_table recursively and then counting total number of True-values in the dp_table through length-wise or diagonal-itenration. 
### here similarly we're doing so but with minor difference :
- difference is that here we;re going to fill the T/F values in the 2-d dp_table through length-iteration or diagonal-iteration while doing this whenever we encountered with True-value we'll increament the count for it.   

- As for longest diagonal ((0,0)(1,1) (2,2)..etc is denoting subproblem having substring of length =1 ) so it means for each & every char in the given string will be considerd as substring in itself and also considered as pallindrome certainly.

- For length = 2  ; we check for j =i+1  //here we check whether adjacent chars of length=2-substring is equal or not. if substring has adjacent-equal chars then fill True else false.

- now any length more than 2 we check first & last chars of substring is equal or not (s[i]==s[j]) and alongwith it we check substrings in-between this equal chars pallindrome or not (recursive_pallindrome(i+1,j-1)->T/F); this i+1 and j-1 also denotes to the relatively smaller-lenthed diagonal (length = curr-length-2) where values we just filled already(solution already computed).

```python
# Palindromic Substrings problem Tabulation(bottom-up) approach [code implementation]:

def countSubstrings(s): # this function specifically used to count total number of pallindromic substring possible  for given original string
    max = 0  # max_counter intialised with 0
    n = len(s) # length of given string 

    dp = [[0]*n for _ in range(n)] #2-d dp_table we construct of dimesnion (n*n)  intialised all cells with 0

    for l in range(1,n+1): # length from 1 upto n
        for i in range(n-l+1): # i from 0 to n-l
            j = i+l - 1  # for each instance of i; j = i+l -1 
            if i==j:  # Longest-diagonal(median-diagonal): this diagonal where length equal to 1 ; means each and every char itself considered as substring 
                #and also considered Pallindrome certainly. 
                dp[i][j] = True
                max+=1  # increasing max counter as whenever we encountered True value is bieng filled
#last chars of curr-substring is equal then further check for substring in-betwwen these equal chars whether they're pallindromic substring or not (dp(i+1,j-1))
#// As we already computed value for substring in-between equal chars also  [that's the values already filled-in at smaller length diagonals ; length = curr-length-2] 
        # if its the case of adjacent equal chars (j==i+1) then no need to check pallindromic substring for further chars in-between. 
            elif s[i]==s[j] and (j==i+1 or dp[i+1][j-1]):  
                dp[i][j] = True
                max +=1  # increasing max counter as whenever we encountered True value is bieng filled
            else:
                dp[i][j] = False
    return max # return total numbers of true values in  the dp_table        

#example usage:
string = 'pqprps'
print(countSubstrings(string)) # should be  8  
st= 'abbabb'
print(countSubstrings(st)) # should be  11     
```
