# Longest pallindromic subsequence problem statement:
### Coding Exercise: Longest Palindromic Subsequence
### Given a string s, find the longest palindromic subsequence's length in s.

### A subsequence is a sequence that can be derived from another sequence by deleting some or no elements without changing the order of the remaining elements.

### Example :

- Input: s = "cccacpc"
- Output: 5
- Explanation: One possible longest palindromic subsequence is "ccccc".

# Longest pallindromic subsequence problem[Tabulation(bottom-up)approach]:
### how this problem diffrent from Longest pallindromic substring problems 


### intuition around approach for solving longest pallindromic subsequence:


### steps to fill the values in the 2-d dp_table here using formulas to count length of pallindromic subsequence for all the cells or these subproblems: 

- we will fill all the values iteratively (using tabulation approach) where iteration will be the lenth-based or diagonal-based(gap-strategy)
- finally after filling-up the dp_table major solution or answer to the problem at hand will be at dp[0][n-1]

# Longest pallindromic subsequence problem[Tabulation(bottom-up)approach] [code implementation]:


```python
# Longest pallindromic subsequence problem[Tabulation(bottom-up)approach(gap-strategy)] [code implementation]:

def longestPalindromeSubseq(s):
    n = len(s)  #length of given string 
    dp = [[0]*n for _ in range(n)] # construncting 2-d dp_table of dimension (n*n) intilised with all cell values equals to zero
    #going to fill dp_table cell values through length-iteration or diagonal iteration [gap-strategy]method 
    for l in range(1, n+1): # length from 1 upto n (upto 6 length)
        for i in range(n-l+1): # i from 0 to n-l
            j = i+l -1 # for each instance of i; j = i+l -1 
            if i==j: # Longest-diagonal(median-diagonal): this diagonal where length equal to 1 ; 
                #means each and every char itself considered as pallindromic subsequence 
                dp[i][j]=1
            elif s[i]==s[j]:  # if first and last chars of subrproblem(subsequence) is equal 
                dp[i][j] = 2 + dp [i+1][j-1]   # then length will be 2 + length of pallindromic-subsequence possible for further subproblem in-between those equal chars.    
            else:  # if first and last chars of subrproblem(subsequence) is NOT equal
                #then we would try to check and calcluate the length of pallindromic-subsequence possible : -[---] (i+1,j) & [---]- (i,j-1)
                dp[i][j] = max(dp [i+1][j], dp[i][j-1])  # going to choose max or longest length from these two pallindromic-subsequneces possibilities.
    return dp[0][n-1]   # major solution/ANS to the problem at hand will be at top-right dp[0][n-1] 
#example usage
st= "pqprps"   #possible longest subsequence here would be 'ppp' 'pqp' and 'prp'    //maximum or longest length pallindrome subseq here is 3
print(longestPalindromeSubseq(st))   # should be 3
```

```python
def longestPalindromeSubseq(s):
    #Write code here
    n = len(s)
    dp = [[0]*n for _ in range(n)]
    for l in range(1, n+1):
        for i in range(n-l+1):
            j = i+l -1
            if i==j:
                dp[i][j]=1
            elif s[i]==s[j]:
                dp[i][j] = 2 + dp [i+1][j-1]
            else:
                dp[i][j] = max(dp [i+1][j], dp[i][j-1])
    return dp[0][n-1]    
```
